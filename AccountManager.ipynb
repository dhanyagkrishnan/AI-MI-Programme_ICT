```python
#Account Information
account_holder="Dhanya Krishnan"
account_number="234567785"
balance=1000.00
currency="Rs"
```

```python
def Deposit_Amount(amount):
 global balance
 if amount<=0:
  print("Deposit amount must be positive");
  return balance
 balance+=amount
 print("Deposit Successfull")
 return balance
```

```python
def Withdrawal_Amount(amount):
 global balance
 if amount <=0:
  print(f"Withdrawal amount not be zero or less than zero")
  return
 elif(amount>balance):
  print(f"Insufficient funds available:{currency}{balance}")
  return
 else:
   balance-=amount
   print("Withdraw successfully")
```

```python
def Check_Balance():
  print(f"Account Holder:{account_holder}")
  print(f"Account Number:{account_number}")
  print(f"Current Balance:{currency}{balance}")
```

```python
Deposit_Amount(100)
Withdrawal_Amount(0)
Check_Balance()
```
