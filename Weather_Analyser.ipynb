```python
#Calculate Average Temperature
def average_temp(data):
  total=0
  count=0
  for day in data:
    total+=sum(data[day]['temperature'])
    count+=len(data[day]['temperature'])

  avg=total/count
  print("Average Temperature:",avg)
  return avg
```

```python
#Calculate Average Temperature
def average_hum(data):
  total=0
  count=0
  for day in data:
    total+=sum(data[day]['humidity'])
    count+=len(data[day]['humidity'])

  avg=total/count
  print("Average humidity:",avg)
  return avg
```

```python
def check_weather(avg_temp):
  if(avg_temp>30):
    print("weather is hot")
  elif (avg_temp<30 and avg_temp>25):
    print("Weather is Pleasant")
  elif(avg_temp<22 and avg_temp>15):
    print('Weather is cold')
  else:
    print("Weather is too Cold")
```

```python
avg_temp=average_temp(weather_data)
avg_hum=average_hum(weather_data)
check_weather(avg_temp)
```
